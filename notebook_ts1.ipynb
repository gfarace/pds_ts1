## Importación de módulos para Jupyter

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pdsmodulos as pds
```

## Función senoidal

```python
def senoidal(vmax, dc, ff, ph, nn, fs):
    ts = 1/fs # tiempo de muestreo
    df = fs/nn # resolución espectral
    
    # grilla de sampleo temporal
    #empieza en 0, termina en (nn-1)*ts y tiene nn números
    #tt es de nn filas x 1 columna
    tt = np.linspace(0, (nn-1)*ts, nn).flatten()
    
    # grilla de sampleo frecuencial
    sen = vmax*np.sin(2*np.pi*ff*tt + ph)+dc
    
    return tt, sen
```

## Gráfica

```python
tt,xx = senoidal(vmax=1, dc=0, ff=1, ph=0, nn=1000, fs=100)

plt.figure(1)
line_hdls = plt.plot(tt, xx)
plt.title('Señal senoidal')
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')

axes_hdl = plt.gca()

plt.show()
```

## Otra señal
### Señal cuadrada

```python
from scipy import signal
import matplotlib.pyplot as plt

inicio = 0
fin = 1
cant = 500
freq = 5

t = np.linspace(inicio, fin, cant, endpoint=False)
yy = signal.square(2 * np.pi * freq * t)

plt.figure(1)
line_hdls = plt.plot(t, yy)
plt.title('Señal cuadrada')
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')

axes_hdl = plt.gca()

plt.show()
```
